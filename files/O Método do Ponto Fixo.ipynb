# O Método do Ponto Fixo

Vamos resolverf a equação $e^x - 4x=0$.

Vamos usar a função de iteração $\phi(x) = e^x/4$.

```julia
# Método do Ponto Fixo
using Printf;
f(x) = exp(x)-4x;
sol_ex = 3.574029561813889e-01;
c_error(x) = abs(x-sol_ex);
phi(x) = 0.25*exp(x);
x = 1.0;
tol = 1.e-16;
error = abs(f(x));
iter = 0;
@printf("%5d   %1.15e\n",iter,x);
while error >tol
    iter += 1;
    x = phi(x);
    error = abs(f(x));
    @printf("%5d   %1.15e\n",iter,x);
end
```

```julia
# Método do Ponto Fixo
using Printf;
using Plots;
f(x) = exp(x)-4x;
sol_ex = 3.574029561813889e-01;
c_error(x) = abs(x-sol_ex);
phi(x) = 0.25*exp(x);
x = 1.0;
tol = 1.e-8;
error = abs(f(x));
iter = 0;
exato = 3.574029561813889e-01;
ERRO = Float64[];
ITER = Int64[];
while error >tol
    
    error = abs(x-exato);
    push!(ERRO,error);
    push!(ITER,iter);
    @printf("%5d   %1.15e\n",iter,error);
    iter += 1;
    x = phi(x);
    
end
```

```julia
plot(ITER,ERRO,markershape = :ltriangle,title="Error Sequence")
```

```julia
plot(ITER,ERRO,markershape = :ltriangle, yaxis=(:log10,(1e-9,1e0)),title="Error Sequence (log scale)")
```

```julia
# Método do Ponto Fixo
using Printf;
using Plots;
f(x) = x^2 + x -6
exato = 2.0;

ϕ1(x) = sqrt(6-x);
ϕ2(x) = 6.0/(x+1);


x = 1.01;
tol = 1.e-12;
#error = abs(f(x));
iter = 0;

ERRO1 = Float64[];
ITER1 = Int64[];
while true
    
    error1 = abs(x-exato);
    push!(ERRO1,error1);
    push!(ITER1,iter);
    @printf("%5d   %18.15e   %18.15e\n",iter,x,error1);
    if error1 < tol
        println("Solution has found.")
        break;
    end
    iter += 1;
    if iter > 1000
        println("  Method has achieved maximum of iterations. Stopping...");
        break;
    end
    x = ϕ1(x);
    
end
```

```julia
plot(ITER1,ERRO1,markershape = :ltriangle, yaxis=(:log10,(1e-17,1e0)),title="Error Sequence (log scale)")
```

```julia
# Método do Ponto Fixo
using Printf;
using Plots;
f(x) = x^2 + x -6
exato = 2.0;

ϕ1(x) = sqrt(6-x);
ϕ2(x) = 6.0/(x+1);
ρ = 0.19;
ϕ3(x) = x - ρ * f(x);

x = 1.01;
tol = 1.e-12;
#error = abs(f(x));
iter = 0;

ERRO3 = Float64[];
ITER3 = Float64[];
while true
    
    error = abs(x-exato);
    push!(ERRO3,error);
    push!(ITER3,iter);
    @printf("%5d   %18.15e   %18.15e\n",iter,x,error);
    if error < tol
        println("Solution has found.")
        break;
    end
    iter += 1;
    if iter > 1000
        println("  Method has achieved maximum of iterations. Stopping...");
        break;
    end
    x = ϕ3(x);
    
end
```

```julia
plot(ITER1,ERRO1,markershape = :ltriangle,label="ϕ1",yaxis=(:log10,(1e-17,1e1)),title="Error Sequence (log scale)")
plot!(ITER2,ERRO2,markershape = :ltriangle,label ="ϕ2",yaxis=(:log10,(1e-17,1e1)), title="Error Sequence (log scale)")
plot!(ITER3,ERRO3,markershape = :ltriangle,label="ϕ3", yaxis=(:log10,(1e-17,1e1)),title="Error Sequence (log scale)")
```
