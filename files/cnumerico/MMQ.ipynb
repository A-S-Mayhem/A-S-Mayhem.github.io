# O Método dos Mínimos Quadrados (MMQ)

Dado um conjunto de pontos do tipo $(x_i,f_i)$ para $i=1, \cdots,n$. Vamos determinar o polinômio de segundo grau que melhor aproxima esses pontos utilizando o método dos mínimos quadrados.

```julia
using Plots

n = 50
x = range(0,1,length = n)
y = 1.5 * x .+ 2 + randn(n) * 0.3
scatter(x,y,legend=false,xlabel="x",ylabel="y")
savefig("mmq.png") 
```

```julia
E(β₀,β₁) = sum((y[i] - β₀ - β₁ * x[i])^2 for i = 1:n)
```

```julia
β₀ = 1.8
β₁ = 2.2

scatter(x,y,leg=false)
plot!(x -> β₀ + β₁ * x, extrema(x)...,c =:green, lw=2)
title!("Soma dos quadrados dos resíduos = $(E(β₀,β₁))")
```

```julia
β₀range = range(1,3,length=100)
β₁range = range(1,3,length=100)
contour(β₀range,β₁range,E,levels=50)
```

```julia
M = [n sum(x); sum(x) sum(x .^2) ]
c = [sum(y);sum( x .* y)]
sol = M \ c
```

```julia
scatter(x,y, leg=false)
plot!(x -> β₀ + β₁ * x, extrema(x)...,c =:green, lw=2)
plot!( x-> sol[1]+sol[2]*x,lw=2,c=:red,xlabel="x",ylabel="y")
savefig("mmq2.png") 
```

```julia
β₀range = range(1,3,length=100)
β₁range = range(1,3,length=100)
contour(β₀range,β₁range,E,levels=50)
scatter!([sol[1]],[sol[2]])
```
