# O Método de Taylor de Ordem $q$

Vamos resolver o problema de valor inicial $$ \left\{ \begin{array}{l} y' = -y + x + 2, \quad [0, 0.3] \\
y(0) = 2\end{array} \right. $$ utilizando o Método de Taylor de ordem 3. Assim, vamos considerar $$y_{n+1}=y_n + h f_n +\frac{h^2}{2} f'_n + \frac{h^3}{3!}f''_n.$$

```julia
using Plots, LinearAlgebra
```

Definição das funções $f,f',f''$

```julia
f(x,y) = -y+x+2
df(x,y) = y - x -1
ddf(x,y) = -y + x + 1
```

Definição do domínio $[x_0,x_f]$ e da condição inicial $y(x_0)=y_0$

```julia
x0 = 0
xf = 0.3
dx = 0.005
y0 = 2
```

Definição do método de Taylor de ordem 3

```julia
ϕ(x,y) = y + dx * f(x,y)  #+dx^2/2 * df(x,y)+dx^3/6 * ddf(x,y)
```

Código que resolve o problema

```julia
x = Float64[]
y = Float64[]
push!(x,x0)
push!(y,y0)
t = x0
yn = y0;
while t <= xf
    global yn = ϕ(t,yn);
    push!(y,yn);
    global t = t + dx;
    push!(x,t)
end
```

Solução Aproximada

```julia
plot(x,y,lw=2)
```

```julia
# Solução Exata
ex(x) = exp(-x) + x + 1
```

Solução Aproximada $\times$ Solução Exata

```julia
plot!(x,ex.(x),lw=2)
```
