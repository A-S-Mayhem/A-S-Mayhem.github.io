# This is only a workaround to provide pagination for a category

## Jekyll plugin pagination v1 only provides pagination for posts

## This method replace the file in _pages for the category

In this case The example will use the "rolling-archive.html" file for the rolling blog posts that create a folder "rollings" (for the category of the same name) in the site

As pagination is not working for this category, we replace the original file for a a file structure with the pagination. We precreate all this pages until the user defined "MAXPAGE"

FOLDERCATEGORY -
  index.htmlpaginationpath
  page-2
     index.html
  page-3
     index.html
  page-4
     index.html
  ...
  
  page-MAXPAGE
     index.html

```python
# Fixed, internal folder of pages to create with jekyll
basefolder = '_pages'

# Params
maxfolder = 100

# The category. In the site, the posts of this categories will be hosted in this path
category = 'rollings'
# Title for the page containing the list of posts
title = "Personal daily events"


# Folder for pagination at the site
paginationpath = 'rolling-archive'

# Require redirect from wordpress?
bredirection = True
pathredirect = "rolling-posts"

# use the number of posts per page from the main settings
#site.paginate
```

```python
for index in range(1,maxfolder+1):
    # Create folder 
    dirCategory = "./" + basefolder + "/" + category + "/"
    if not os.path.exists(dirCategory):
        os.mkdir(dirCategory)

    md_file = open("./" + basefolder + "/" + category + "/" + "page" + str(index) + ".html", "w")
        

    md_file.write("---")
    md_file.write("\n")

    md_file.write("index: " + str(index))
    md_file.write("\n")

    md_file.write("layout: archive")
    md_file.write("\n")
        
    if index == 1:
        md_file.write("permalink: /" + paginationpath + "/index.html")
    else:    
        md_file.write("permalink: /" + paginationpath + "/page" + str(index) + "/index.html")
    
    md_file.write("\n")
    md_file.write("title: " + title)
    md_file.write("\n")
    md_file.write("author_profile: true")
    md_file.write("\n")

    if bredirection:
        md_file.write("redirect_from:")
        md_file.write("\n")
        
        if index == 1:
            md_file.write("  - /wordpress/" + pathredirect + "/")
            
        else:
            md_file.write("  - /wordpress/" + pathredirect + "/page" + str(index) + "/")
        
        md_file.write("\n")

    md_file.write("---")
    md_file.write("\n")
    md_file.write("\n")

    md_file.write("{% include base_path %}")
    md_file.write("\n")
    md_file.write("{% capture written_year %}'None'{% endcapture %}")
    md_file.write("\n")
    md_file.write("{% assign sortedposts = site." + category + " | sort: 'date' | reverse %}")
    md_file.write("\n")

    # Add here a selector of the posts for this page
    # using index, maxfolder, 

    md_file.write("\n")
    md_file.write("{% assign is_last_partial = 0 %}")
    md_file.write("\n")
    md_file.write("{% assign have_previus_page = 0 %}")
    md_file.write("\n")
    md_file.write("{% assign have_next_page = 0 %}")
    md_file.write("\n")
    md_file.write("{% assign have_valid_page = 0 %}")
    md_file.write("\n")

    md_file.write("\n")
    md_file.write("\n")
    md_file.write("{% assign posts_count = sortedposts | size %}")
    md_file.write("\n")

    md_file.write("\n")
    md_file.write("\n")
    md_file.write("{% assign internal_mod_pag = 1000 %}")
    md_file.write("\n")
    md_file.write('{% if site.paginate > 0 %}')
    md_file.write("\n")
    md_file.write("      {% assign internal_mod_pag = site.paginate %}")
    md_file.write("\n")
    md_file.write("{% endif %}")
    md_file.write("\n")



    md_file.write("\n")
    md_file.write("\n")
    md_file.write("{% assign mod_pag = posts_count | modulo:internal_mod_pag %}")
    md_file.write("\n")
    md_file.write('{% if mod_pag > 0 %}')
    md_file.write("\n")
    md_file.write("      {% assign is_last_partial = 1 %}")
    md_file.write("\n")
    md_file.write("{% endif %}")
    md_file.write("\n")
    md_file.write("{% assign pages_count = posts_count | divided_by:internal_mod_pag | plus:is_last_partial %}")
    md_file.write("\n")

    md_file.write("\n")
    md_file.write("\n")
    md_file.write("{% assign next_pag = page.index | plus:1 %}")
    md_file.write("\n")

    md_file.write("\n")
    md_file.write("\n")
    md_file.write("{% assign previus_pag = page.index | minus:1 %}")
    md_file.write("\n")


    md_file.write("\n")
    md_file.write("{% if next_pag <= pages_count %}")
    md_file.write("\n")
    md_file.write("      {% assign have_next_page = 1 %}")
    md_file.write("\n")
    md_file.write("{% endif %}")
    md_file.write("\n")

    md_file.write("\n")
    md_file.write("{% if previus_pag > 0 %}")
    md_file.write("\n")
    md_file.write("      {% assign have_previus_page = 1 %}")
    md_file.write("\n")
    md_file.write("{% endif %}")
    md_file.write("\n")


    md_file.write("\n")
    md_file.write("\n")
    md_file.write("{% assign internaltimes = page.index | minus:1 %}")
    md_file.write("\n")
    md_file.write("{% assign OFFSET = internal_mod_pag | times:internaltimes %}")
    md_file.write("\n")


    md_file.write("\n")
    md_file.write("{% if page.index <= pages_count %}")
    md_file.write("\n")
    md_file.write("      {% assign have_valid_page = 1 %}")
    md_file.write("\n")
    md_file.write("{% endif %}")
    md_file.write("\n")
    md_file.write("\n")
    md_file.write("\n")

    # If valid page
    md_file.write("\n")
    md_file.write("{% if have_valid_page > 0 %}")
    md_file.write("\n")
    md_file.write("      {% for post in sortedposts limit:internal_mod_pag offset:OFFSET %}")
    md_file.write("\n")
    md_file.write("        {% capture year %}{{ post.date | date: '%Y' }}{% endcapture %}")
    md_file.write("\n")
    md_file.write("        {% if year != written_year %}")
    md_file.write("\n")
    md_file.write('          <h2 id="{{ year | slugify }}" class="archive__subtitle">{{ year }}</h2>')
    md_file.write("\n")
    md_file.write("          {% capture written_year %}{{ year }}{% endcapture %}")
    md_file.write("\n")
    md_file.write("        {% endif %}")
    md_file.write("\n")
    md_file.write("        {% include archive-single.html %}")
    md_file.write("\n")
    md_file.write("      {% endfor %}")
    md_file.write("\n")
    md_file.write("{% endif %}")
    md_file.write("\n")
    md_file.write("\n")



    md_file.write("\n")
    md_file.write("\n")
    md_file.write("{% if pages_count > 1 %}")
    md_file.write("\n")
    md_file.write('<nav class="pagination">')
    md_file.write("\n")
    md_file.write("  <ul>")
    md_file.write("\n")
    md_file.write("    {% comment %} Link for previous page {% endcomment %}")
    md_file.write("\n")
    md_file.write("    {% if have_previus_page > 0 %}")
    md_file.write("\n")
    md_file.write("      {% if previus_pag == 1 %}")
    md_file.write("\n")
    md_file.write('        <li><a href="{{ base_path }}/' + paginationpath + '/">{{ site.data.ui-text[site.locale].pagination_previous | default: "Previous" }}</a></li>')
    md_file.write("\n")
    md_file.write("      {% else %}")
    md_file.write("\n")
    md_file.write('        <li><a href="{{ base_path }}/' +  paginationpath + '/page{{ previus_pag }}/">{{ site.data.ui-text[site.locale].pagination_previous | default: "Previous" }}</a></li>')
    md_file.write("\n")
    md_file.write("      {% endif %}")
    md_file.write("\n")
    md_file.write("    {% else %}")
    md_file.write("\n")
    md_file.write('      <li><a href="#" class="disabled"><span aria-hidden="true">{{ site.data.ui-text[site.locale].pagination_previous | default: "Previous" }}</span></a></li>')
    md_file.write("\n")
    md_file.write("    {% endif %}")
    md_file.write("\n")
    md_file.write("    {% comment %} First page {% endcomment %}")
    md_file.write("\n")
    md_file.write("    {% if page.index == 1 %}")
    md_file.write("\n")
    md_file.write('      <li><a href="#" class="disabled current">1</a></li>')
    md_file.write("\n")
    md_file.write("    {% else %}")
    md_file.write("\n")
    md_file.write('      <li><a href="{{ base_path }}/' + paginationpath + '/">1</a></li>')
    md_file.write("\n")
    md_file.write("    {% endif %}")
    md_file.write("\n")
    md_file.write("    {% assign page_start = 2 %}")
    md_file.write("\n")
    md_file.write("    {% if page.index > 4 %}")
    md_file.write("\n")
    md_file.write("      {% assign page_start = page.index | minus: 2 %}")
    md_file.write("\n")
    md_file.write("      {% comment %} Ellipsis for truncated links {% endcomment %}")
    md_file.write("\n")
    md_file.write('      <li><a href="#" class="disabled">&hellip;</a></li>')
    md_file.write("\n")
    md_file.write("    {% endif %}")
    md_file.write("\n")
    md_file.write("\n")
    md_file.write("    {% assign page_end = pages_count | minus: 1 %}")
    md_file.write("\n")
    md_file.write("    {% assign pages_to_end = pages_count | minus: page.index %}")
    md_file.write("\n")
    md_file.write("    {% if pages_to_end > 4 %}")
    md_file.write("\n")
    md_file.write("      {% assign page_end = page.index | plus: 2 %}")
    md_file.write("\n")
    md_file.write("    {% endif %}")
    md_file.write("\n")
    md_file.write("\n")
    md_file.write("    {% for index in (page_start..page_end) %}")
    md_file.write("\n")
    md_file.write("      {% if index == page.index %}")
    md_file.write("\n")
    md_file.write('        <li><a href="{{ base_path }}/' + paginationpath + '/page{{ index }}/" class="disabled current">{{ index }}</a></li>')
    md_file.write("\n")
    md_file.write("      {% else %}")
    md_file.write("\n")
    md_file.write("        {% comment %} Distance from current page and this link {% endcomment %}")
    md_file.write("\n")
    md_file.write("        {% assign dist = page.index | minus: index %}")
    md_file.write("\n")
    md_file.write("        {% if dist < 0 %}")
    md_file.write("\n")
    md_file.write("          {% comment %} Distance must be a positive value {% endcomment %}")
    md_file.write("\n")
    md_file.write("          {% assign dist = 0 | minus: dist %}")
    md_file.write("\n")
    md_file.write("        {% endif %}")
    md_file.write("\n")
    md_file.write('        <li><a href="{{ base_path }}/' +  paginationpath + '/page{{ index }}/">{{ index }}</a></li>')
    md_file.write("\n")
    md_file.write("      {% endif %}")
    md_file.write("\n")
    md_file.write("    {% endfor %}")
    md_file.write("\n")
    md_file.write("    {% comment %} Ellipsis for truncated links {% endcomment %}")
    md_file.write("\n")
    md_file.write("    {% if pages_to_end > 3 %}")
    md_file.write("\n")
    md_file.write('      <li><a href="#" class="disabled">&hellip;</a></li>')
    md_file.write("\n")
    md_file.write("    {% endif %}")
    md_file.write("\n")
    md_file.write('    {% if page.index == pages_count %}')
    md_file.write("\n")
    md_file.write('      <li><a href="#" class="disabled current">{{ page.index }}</a></li>')
    md_file.write("\n")
    md_file.write("    {% else %}")
    md_file.write("\n")
    md_file.write('      <li><a href="{{ base_path }}/' + paginationpath + '/page{{ pages_count }}/">{{ pages_count }}</a></li>')
    md_file.write("\n")
    md_file.write("    {% endif %}")
    md_file.write("\n")
    md_file.write("\n")
    md_file.write("\n")
    md_file.write("    {% comment %} Link next page {% endcomment %}")
    md_file.write("\n")
    md_file.write("    {% if have_next_page > 0 %}")
    md_file.write("\n")
    md_file.write('      <li><a href="{{ base_path }}/' +  paginationpath +  '/page{{ next_pag }}/">{{ site.data.ui-text[site.locale].pagination_next | default: "Next" }}</a></li>')
    md_file.write("\n")
    md_file.write("    {% else %}")
    md_file.write("\n")
    md_file.write('      <li><a href="#" class="disabled"><span aria-hidden="true">{{ site.data.ui-text[site.locale].pagination_next | default: "Next" }}</span></a></li>')
    md_file.write("\n")
    md_file.write("    {% endif %}")
    md_file.write("\n")
    md_file.write("  </ul>")
    md_file.write("\n")
    md_file.write("</nav>")
    md_file.write("\n")
    md_file.write("{% endif %}")
    md_file.write("\n")


    md_file.close()
```
