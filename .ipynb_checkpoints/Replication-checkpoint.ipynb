# STEP 1: Inferring Routes

To infer routes, we use two novel sources of data: the pairs of municipalities of 1872 in the trade network and the pairs of municipalities of 1872 in the post-office network. The information is in two shape files: \verb|rotas_terrestres_entresedes1873_amc_contig.shp| and \verb|rotas_correios_entresedes_1863_terrestres_amc_contig.shp|, respectively. Although the data refers to municipalities of 1872, the data is from official reports published in 1863 in the case of post-office, and in 1873 in the case of trade routes. For each route we also have information on the actual distance trailed between municipalities. However, we don't know exactly the starting and ending point of the routes so we assume that routes link the administrative centres -- which we refer to as city halls --  of two municipalities.

We use the shape file \verb|sede_municipal_1872.shp| containing the location of the city halls of each municipality in 1872 and a list with pairs of contiguous municipalities in the file \verb|malha.plist_withgoogledist.csv| to create a list of all possible routes.\footnote{Possible and contiguous. We assume that neither post-offices nor traders would cross a municipality without stopping. We also assume that they stop at the city hall.} Using the script \verb|combinePaths.py|, we add to each potential routes information about whether it is in the trade network, in the post-office network, the effective distance reported in the official documents and information about how often post-office routes where trailed by month. The code provides the following information. There are 1766 possible contiguous routes out of which 924 are trade routes and 477 are post-office routes. Most of the post-office routes (315) are also trade routes. For the routes serving both, on average, trade documents reports effective distances smaller than post-office documents, the average difference being 13 kilometers.\footnote{Recall that the documents are 10 years apart such that these differences may result from imputation errors, change in route or change in measurement techniques. Another potential source of mismatch is that distances where measured in leagues, which are rather inaccurate.} We also compare the reported distances with the shortest possible distance between two municipalities. There are 63 routes whose post-offices report distances shorter than haversine distance. This number is 36 for the trade report.

With the information about all possible routes stored in \verb|routes.dta|, we apply a least-cost path method using two cost measures. The main cost measure is stored in the raster file \verb|Ozak_dist_BRA.tif|. It contains information about the Human Mobility Index created by \cite{ozak2010voyage}. The script \verb|create_shapeHMI.py| produces the shape file \verb|HMI_paths.shp| with all possible routes and the information mentioned previously. The code is based in the the \verb|georasters| library provided by \"Omer \"Ozak, which runs \verb|skimage| in the background.The same process is performed by the file \verb|create_shapeRUD.py| using ruggedness as cost. It produces the shape file \verb|RUD_paths.shp|. The raster with ruggedness information is named \verb|rugBRA.tif|.

This step is time consuming as the method involves the search over grids of size 1x1 kilometers in the HMI and 250x250 meters in the case of ruggedness.  Computing the least-cost paths using the HMI rasters using 8 cores with an Intel Core i7-9700 3.00Ghz processor takes approximately 30 minutes. In the case of ruggedness, since grids a smaller, matrix sizes are extremely large and computation too long.\footnote{There are matrices that require 2.62 GiB to be handled.} Then, we down-sampled the ruggedness raster to the same resolution as the HMI raster averaging the information. Since the ruggedness raster is more granular, it takes longer. The code is written such that all minus one cores in the computer are used. Change the number of workers in the code if you want to alter this feature.

\paragraph{Summary:} Running the following lines produces . The first line combines information about different types of routes. The second creates a shape file with the least-cost path using HMI as cost, whereas the third does the same using ruggedness as cost. The code should be able to identify the location of the original script to access raw data and save output files. If any error is raised about the location, adjust the working directory directly into the files.

```python
hmishape.crs = 'EPSG:4367' 
hmimeters = hmishape.copy()
hmimeters = hmimeters.to_crs('ESRI:102032')
hmimeters['length_HMI'] = hmimeters.geometry.length/1e3

rugshape.crs = 'EPSG:4367' 
rugmeters = rugshape.copy()
rugmeters = rugmeters.to_crs('ESRI:102032')
rugmeters['length_RUG'] = rugmeters.geometry.length/1e3

dist = hmimeters.merge(rugmeters[['routeID', 'length_RUG']], on='routeID', how='inner')
dist['efeLength'] = dist['dist_efeti']
dist.loc[dist.trade == 1, 'efeLength'] = dist['dist_efe_1']
```

```python
import statsmodels.api as sm

regdata = dist[['efeLength', 'length_HMI', 'length_RUG']]
regdata['constant'] = 1

modHMI = sm.OLS(regdata.dropna(subset=['efeLength', 'length_HMI']).length_HMI, regdata.dropna(subset=['efeLength', 'length_HMI']).loc[:,['constant', 'efeLength']])
resHMI = modHMI.fit()

modRUG = sm.OLS(regdata.dropna(subset=['efeLength', 'length_RUG']).length_RUG, regdata.dropna(subset=['efeLength', 'length_RUG']).loc[:,['constant', 'efeLength']])
resRUG = modRUG.fit()

r2Rug = round(resRUG.rsquared, 2)
betarug = round(resRUG.params[1], 2)
r2HMI = round(resHMI.rsquared, 2)
betaHMI = round(resHMI.params[1], 2)
```

```python
mail = dist.loc[(dist.mail == 1) & (dist.trade!= 1)]
trade = dist.loc[(dist.mail != 1) & (dist.trade== 1)]
both = dist.loc[(dist.mail == 1) & (dist.trade== 1)]
maxdist = max(dist.dist_efeti.max(), dist.dist_efe_1.max())
line45 = np.linspace(0, maxdist, 1000)

dist['hmisse'] = (dist['efeLength']-dist['length_HMI'])**2
dist['rugsse'] = (dist['efeLength']-dist['length_RUG'])**2

hmicorr = round(dist[['efeLength', 'length_HMI']].corr().iloc[0,1], 2)
rugcorr = round(dist[['efeLength', 'length_RUG']].corr().iloc[0,1], 2)
hmiSSE = round(sum(dist['hmisse'].dropna()), 0)
rugSSE = round(sum(dist['rugsse'].dropna()), 0)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
ax1.scatter(mail.dist_efeti, mail.length_HMI,c=mailColor, alpha=0.4, label='Post Office Only')
ax1.scatter(trade.dist_efe_1, trade.length_HMI,c=tradeColor, alpha=0.4, label='Trade Only')
ax1.scatter(both.dist_efe_1, both.length_HMI,c=bothColor, alpha=0.4, label='Both')
ax1.plot(line45, line45, c='r', ls='--', label=r'$45^o$ Line')
ax1.set_xlim(0, maxdist), ax1.set_ylim(0, maxdist)
ax1.set_xlabel('Effective Distance\n (a) HMI'), ax1.set_ylabel('Estimated Distance')
#ax1.set_title('Correlation:'+str(hmicorr)+' | SSE: '+str(hmiSSE))
ax1.set_title('OLS Regression - Slope:'+str(betaHMI)+' | R$^2$: '+str(r2HMI))

ax1.legend()

ax2.scatter(mail.dist_efeti, mail.length_RUG,c=mailColor, alpha=0.4)
ax2.scatter(trade.dist_efe_1, trade.length_RUG,c=tradeColor, alpha=0.4)
ax2.scatter(both.dist_efe_1, both.length_RUG,c=bothColor, alpha=0.4)
ax2.plot(line45, line45, c='r', ls='--', label=r'$45^o$ Line')
ax2.set_xlim(0, maxdist), ax1.set_ylim(0, maxdist), 
ax2.set_xlabel('Effective Distance\n (b) Ruggedness'), ax2.set_ylabel('Estimated Distance')
# ax2.set_title('Correlation:'+str(rugcorr)+' | SSE: '+str(rugSSE))
ax2.set_title('OLS Regression - Slope:'+str(betarug)+' | R$^2$: '+str(r2Rug))
fig.tight_layout()
```
